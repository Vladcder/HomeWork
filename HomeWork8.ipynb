**Работа с API**

```python
import requests
import pandas as pd
```

```python
api_base_url = 'https://www.anapioficeandfire.com/api'
```

Информация о книгах, домах и домах с девизом

```python
def get_info(infoType, hasTitles):
  param=''
  if hasTitles:
    list=requests.get(f'{api_base_url}/{infoType}?hasTitles=true&pageSize=50').json()
  else:
    list=requests.get(f'{api_base_url}/{infoType}?pageSize=50').json()
  return list
```

```python
def get_all_info(infoType, findWithTitles):
  list=[]
  page=1
  while (True):
    url=f'{api_base_url}/{infoType}/'+str(page)
    try:
          data=requests.get(url).json()
          if findWithTitles:
            if len(data.get('titles')):
              list.append(data)
          else:
            list.append(data)
          page+=1
    except ValueError:
           return list
  return list
```

```python
all_houses=get_all_info('houses', False)
all_books=get_all_info('books', False)
all_houses_with_titles=get_all_info('houses',True)
```
