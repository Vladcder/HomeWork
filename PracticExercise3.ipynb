Самое популярное имя

```python
df_names=df[["Name","Sex"]]
first_name_array=np.empty(len(df_names), dtype=object)
```

```python
def getFirstNames(names):
  first_name_array=np.empty(len(names), dtype=object)
  i=0
  for name in names["Name"]:
    if (("(" in name) and (names["Sex"].iloc[i]=='female')):
      fl=name.split("(")
      first_name_array[i]=fl[1].split()[0]
      i+=1
    else:
      fl = name.split(".")
      first_name_array[i]=fl[1].split()[0]
      i+=1
  return first_name_array


```

```python
first_name_array=getFirstNames(df_names)
df["FirstName"]=first_name_array
df.head()
```

```python
popular_female_name=df.groupby("Sex")["FirstName"].value_counts()['female']
print("Самое популярное женское имя "+str(popular_female_name.index[0])+". Оно встречается "+ str(popular_female_name.iloc[0]))
popular_male_name=df.groupby("Sex")["FirstName"].value_counts()['male']
print("Самое популярное мужское имя "+str(popular_male_name.index[0])+". Оно встречается "+ str(popular_male_name.iloc[0]))
```
